```python
train=pd.read_csv("./train.csv")
test=pd.read_csv("./test.csv")
train.head(4)
```

```python
#creating new column of log_revenue because hist representation of simple revenue column is not good represent
fig,ax=plt.subplots(figsize=(16,6))
plt.subplot(1,2,1)
plt.title('Distribution of log transformed revenue')
#creating new column of log transformation of original distribution
train['log_revenue']=np.log1p(train['revenue'])
train.log_revenue.hist()
```

```python
#Task 1:Analysis relation between film revenue and budget
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.scatterplot(train['budget'],train['revenue'])
plt.title('revenue vs budget');
plt.subplot(1,2,2)
sns.scatterplot(np.log1p(train['budget']),train['log_revenue'])
plt.title('logrevenue vs log_budget');
```

```python
train['log_budget']=np.log1p(train['budget'])
test['log_budget']=np.log1p(test['budget'])

```

```python
#task 2: does having an official homepage affect revenue?
train['homepage'].value_counts().head(10)
train['has_homepage']=0
train.loc[train['homepage'].isnull() == False ,'has_homepage']=1

test['has_homepage']=0
test.loc[test['homepage'].isnull() == False, 'has_homepage' ]=1

sns.catplot(x='has_homepage',y='revenue',data= train)
plt.title('Revenue for films with and without homepage');
#so,it affect 
```
